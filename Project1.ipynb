```python
# Exploring Dataset
import pandas as pd
import numpy as np
df = pd.read_csv("Titanic-Dataset.csv")
print(df.head())
print(df.info())
print(df.describe())
```

```python
#Missing Values
MV = df.isnull()
print(MV)
```

```python
#Filtering Data Based on Missing Values
d = pd.read_csv("Titanic-Dataset.csv")
bool_series = pd.isnull(d["Age"])
missing_age_data = d[bool_series]
print(missing_age_data)
```

```python
# Converting Categorical Features into numerical
df.info()
from sklearn import preprocessing
label_encoder1= preprocessing.LabelEncoder()
label_encoder1.fit(df["Sex"])
label_encoder2 = preprocessing.LabelEncoder()
label_encoder2.fit(df["Embarked"])
```

```python
#Normalize/standardize the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[["Age","Fare"]] = scaler.fit_transform(df[["Age", "Fare"]])

print(df[["Age","Fare"]])

from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
df[["Age","Fare"]] = mms.fit_transform(df[["Age","Fare"]])
print(df[["Age","Fare"]])

```
