```python
import pandas as pd
import numpy as np
df = pd.read_csv("Titanic-Dataset.csv")
print(df.head())
print(df.info())
df.describe(include='all')


```

```python
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
for col in df.select_dtypes(include='number'):
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
    for col in df.select_dtypes(include='number'):
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()




```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df.select_dtypes(include='number'))
plt.figure(figsize=(10,6))

```

```python
#The Titanic dataset contains 891 passenger records with features such as age, sex, class, fare, and survival status. The data has notable missing values, mainly in the Age, Cabin, and Embarked columns, where Cabin is almost entirely missing and typically removed. The distribution analysis shows that most passengers were young adults aged 2040, while the Fare variable is highly right-skewed, with a few passengers paying very high ticket prices. Most passengers traveled alone, as indicated by low SibSp and Parch values, though some outliers exist due to large families. Boxplots reveal outliers in Age, Fare, and family-related features. Correlation analysis highlights a strong negative relationship between Fare and Pclass, confirming that higher-class passengers paid more. Overall, the EDA suggests that passenger class, gender, age, and fare likely influence survival outcomes, while missing values and skewed distributions must be handled carefully before model building.
```
